```python
plt.scatter("fixed acidity" , "volatile acidity" , data = df, color = "#660303")
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.title("Fixed Acidity and Volatile Acidity Comparision")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter("fixed acidity" , "volatile acidity" , data = df, color = "#660303")
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.title("Fixed Acidity and Volatile Acidity Comparision")

plt.subplot(1,2,2)
plt.scatter("citric acid" , "residual sugar" , data = df, color = "#100140")
plt.xlabel("Citric Acid")
plt.ylabel("Residual Sugar")
plt.title("Citric Acidity and Residual Sugar Comparision")


plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter("chlorides" , "density" , data = df, color = "#40012f")
plt.xlabel("Chlorides")
plt.ylabel("Density")
plt.title("Chlorides and Density Comparision")


plt.subplot(1,2,2)
plt.scatter("free sulfur dioxide" , "total sulfur dioxide" , data = df, color = "#19ba0d")
plt.xlabel("Free Sulfur Dioxide")
plt.ylabel("Total Sulfur Dioxide")
plt.title("Free Sulfur Dioxide and Total Sulfur Dioxide Comparision")
```
