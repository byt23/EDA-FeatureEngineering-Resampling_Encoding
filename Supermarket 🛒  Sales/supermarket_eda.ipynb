```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Supermarket Sales.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize = (25,20))

plt.subplot(1,2,1)
sns.barplot(x = "Product line" , y = "gross income" , data = df, hue = "Gender", palette = "Set1" )
plt.xlabel("Product Line" , fontsize = 30)
plt.ylabel("Gross Income", fontsize = 30)
plt.title("Product line and Gross Income Comparision Table", fontsize = 25)


plt.subplot(1,2,2)
sns.barplot(x = "Product line" , y = "Rating" , data = df , hue = "Gender"  ,palette = "Set2" )
plt.xlabel("Product Line", fontsize = 30)
plt.ylabel("Rating", fontsize = 30)
plt.title("Product Line and Rating Comparision Table", fontsize = 25)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (25,20))

plt.subplot(1,2,1)
sns.barplot(y = "Product line" , x = "gross income" , data = df, hue = "Gender", palette = "Set1" )
plt.ylabel("Product Line" , fontsize = 30)
plt.xlabel("Gross Income", fontsize = 30)
plt.title("Product line and Gross Income Comparision Table", fontsize = 25)


plt.subplot(1,2,2)
sns.barplot(y = "Product line" , x = "Rating" , data = df , hue = "Gender"  ,palette = "Set2" )
plt.ylabel("Product Line", fontsize = 30)
plt.xlabel("Rating", fontsize = 30)
plt.title("Product Line and Rating Comparision Table", fontsize = 25)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,16))
sns.set_style("whitegrid")
sns.boxplot(x = "Customer type" , y = "gross income" , data = df, hue = "Gender", palette = "dark" , legend=True)
plt.xlabel("Customer Type", fontsize = 30, color = "Blue")
plt.ylabel("Gross Income", fontsize = 30, color = "Blue")
plt.title("Customer type and Gross Income Graph by Gender", fontsize = 30, color = "Blue")
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(x="Customer type",y="Sales", data = df, palette = "Set2" , legend = True, hue = "Gender")
plt.title("Customer Type and Sales Graphs by Customer Types", fontsize = 35)
plt.xlabel("Customer Type", fontsize = 30)
plt.ylabel("Sales" , fontsize = 30)
plt.tight_layout()
plt.show()
```
