```python
df.groupby(["Category"]).std(numeric_only = True)["Price"]
```

```python
plt.figure(figsize =(20,5))
colors = ["red","green","yellow","pink","purple","orange","blue","magenta","brown","Cyan","Indigo","Black","Gray"]
fig = df.groupby(["Category"]).std(numeric_only=True)["Price"].plot(kind = "bar",  color = colors)
fig.set_xticklabels(fig.get_xticklabels(), rotation=90)
plt.show(fig)
```

```python
df.groupby(["Category"])["Price"].mean()
```

```python
df = df[df["Category"] != str(1.9)]
df = df.reset_index(drop=True)
```

```python
df.Category.unique()
```

```python
plt.figure(figsize=(20,5))
colors = ["red","green","yellow","pink","purple","orange","blue","magenta","brown","Cyan","Indigo","Black","Gray"]
fig = df.groupby(["Category"])["Rating"].mean().plot(kind = "bar", title = "Mean Of Categories", color = colors)
fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize = (15,15))

mylabels = df.Category.value_counts().index

ax1.pie(df.Category.value_counts(), labels = mylabels, autopct="%1.1f%%")

plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = df.groupby(["Category"])["Installs"].count().sort_values(ascending = False).head(15).plot(kind = "bar", title = "Top 15 App Categories")
fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
plt.show(fig)
```
