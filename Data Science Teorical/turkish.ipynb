Makine Öğrenmesinde Keşifçi Veri Analizini (EDA) Anlamak 
Gerçek dünya verileriyle çalışırken, doğrudan makine öğrenmesi modelleri oluşturmaya atlamak yaygın bir hatadır. Herhangi bir özellik mühendisliği uygulamadan veya bir modeli eğitmeden önce, Keşifçi Veri Analizi (EDA) kullanarak veriyi anlamamız gerekir.
EDA Neden Önemlidir? 
•	Temelde yatan kalıpları keşfetmemize ve anormallikleri (aykırı değerler veya hatalar gibi) tespit etmemize yardımcı olur.
•	Eksik değerleri ele alma ve veri kodlama (encoding) tekniklerini seçme konusundaki kararlarımıza rehberlik eder.
•	Farklı özellikler ile hedef değişken arasındaki ilişkileri ortaya çıkarır.
EDA'daki Temel Adımlar Veriyi keşfetmek için her biri farklı hedeflere uygun çok sayıda teknik vardır.
1. Betimsel İstatistikler (Descriptive Statistics) Veri setinin temel şeklini anlamak için hızlı bir matematiksel özetini (ortalama, minimum, maksimum ve standart sapma gibi) almak.
•	Örnek: 
Python
import pandas as pd
# Sayısal sütunlar için özet istatistikleri görüntüle
print(df.describe())
# Veri tiplerini ve eksik değerleri kontrol et
print(df.info())
📌 Ne Zaman Kullanılır: Veri setini ilk yüklediğinizde ve üst düzey bir genel bakışa ihtiyaç duyduğunuzda.
2. Tek Değişkenli Analiz (Univariate Analysis - Tek Bir Değişkeni Analiz Etme) Dağılımını, yayılımını veya frekansını anlamak için her seferinde tek bir özelliği incelemek. Sayısal verilerdeki çarpıklığı (skewness) veya kategorik verilerdeki sınıf dengesizliklerini (class imbalances) belirlemek için en ideal yöntemdir.
•	Örnek (Sayısal Dağılım): 
Python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Age'], kde=True)
plt.show()
📌 Ne Zaman Kullanılır: Tek bir özelliğin şeklini ve yayılımını görmek istediğinizde.
3. İki Değişkenli Analiz (Bivariate Analysis - İki Değişkeni Analiz Etme) Genellikle bir yordayıcı (predictor) değişkeni hedef değişkenle karşılaştırarak iki özellik arasındaki ilişkiye bakmak. Eğilimleri ve korelasyonları belirlemeye yardımcı olur.
•	Örnek (Sayısal vs. Sayısal): 
Python
sns.scatterplot(x='Age', y='Income', data=df)
plt.show()
📌 Ne Zaman Kullanılır: Bir özelliğin diğerini nasıl etkilediğini veya onunla nasıl ilişkili olduğunu görmek istediğinizde.
4. Aykırı Değer Tespiti (Outlier Detection) Diğer gözlemlerden önemli ölçüde sapan aşırı uç değerleri belirlemek. Aykırı değerler model performansını (özellikle doğrusal modelleri) ciddi şekilde etkileyebilir ve özet istatistikleri bozabilir.
•	Örnek (Kutu Grafiği - Boxplot Kullanımı): 
Python
sns.boxplot(x=df['Salary'])
plt.show()
📌 Ne Zaman Kullanılır: Sayısal sütunlardaki anormallikleri veya olası veri girişi hatalarını kontrol ederken.
5. Korelasyon Analizi (Correlation Analysis) Sayısal özelliklerin birbirleriyle ne kadar güçlü ilişkili olduğunu ölçmek. Yüksek düzeyde ilişkili özellikler sorunlara (çoklu doğrusal bağlantı) neden olabilir ve genellikle gereksiz (tekrarlayan) bilgi sağlar.
•	Örnek: 
Python
# Korelasyon matrisini hesapla ve çiz
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
📌 Ne Zaman Kullanılır: Model için hangi sayısal özelliklerin en önemli ve farklı olduğunu seçerken.
Özet 
Analiz Türü	Amaç	Yaygın Yöntemler / Grafikler
Betimsel İstatistikler	Hızlı bir matematiksel özet elde etmek	describe(), info()
Tek Değişkenli Analiz	Tek bir özelliğin dağılımını anlamak	Histogramlar, Çubuk Grafikleri, Sayım Grafikleri
İki Değişkenli Analiz	İki özellik arasındaki ilişkileri bulmak	Dağılım Grafikleri, Kutu Grafikleri (Kategorik vs Sayısal)
Aykırı Değer Tespiti	Aşırı, sıra dışı değerleri bulmak	Kutu Grafikleri, Z-Skoru, IQR Yöntemi
Korelasyon	Özelliklerin birlikte nasıl hareket ettiğini görmek	Korelasyon Matrisi, Isı Haritası (Heatmap)
        
Son İpucu: Eksik verileri nasıl ele alacağınıza veya hangi kodlamayı kullanacağınıza karar vermeden önce, verinin hikayesini anlatmasına izin vermek için her zaman EDA yapın. EDA'yı atlamak gözü kapalı araba kullanmaya benzer—ilerleyebilirsiniz, ama eninde sonunda kaza yaparsınız! 
 
Özellik Mühendisliğinde Eksik Değerleri (Missing Values) Anlamak 
Gerçek dünya verileriyle çalışırken eksik değerler yaygın bir sorundur. Özellik mühendisliğinde, eksik değerleri doğru şekilde ele almak çok önemlidir çünkü makine öğrenmesi modellerinin performansını etkileyebilirler.
Eksik Veri Türleri Eksik değerler temel olarak üç türe ayrılabilir: 
1. Tamamen Rastgele Eksik (MCAR - Missing Completely at Random) Eksik değerler rastgele oluşur ve veri setindeki diğer hiçbir değişkenle ilişkisi yoktur.
•	Örnek: Bir kişinin ankette yaşını doldurmayı unutması.
•	Etkisi: Hiçbir yanlılık (bias) oluşturmaz ve eksik değerleri silmek veya ortalama/medyan gibi basit bir yöntemle doldurmak genellikle sorun yaratmaz.
2. Rastgele Eksik (MAR - Missing at Random) Eksik değerler gözlemlenen diğer değişkenlerle ilişkilidir ancak eksik verinin kendisiyle ilişkili değildir.
•	Örnek: Tıbbi bir veri setinde erkeklerin kilo hakkındaki soruyu atlama olasılığı daha yüksektir, ancak her cinsiyetin kendi içinde kilo rastgele eksiktir.
•	Etkisi: Eksik değerleri doldurmak için regresyon veya ilgili özellikleri kullanmak gibi daha gelişmiş doldurma yöntemleri faydalı olabilir.
3. Rastgele Olmayan Eksik (MNAR - Missing Not at Random) Eksik değerler gözlemlenmeyen verinin kendisine bağlıdır, yani eksikliğin nedeni rastgele değildir.
•	Örnek: Yüksek gelire sahip kişiler ankette maaşlarını bildirmemeyi tercih edebilirler.
•	Etkisi: Bu durum yanlılık yaratabilir ve MNAR'ı ele almak genellikle alan bilgisi (domain knowledge) veya eksikliği ayrı ayrı modellemek gibi özel teknikler gerektirir.
Özellik Mühendisliğinde Eksik Değerleri İşleme Eksik değerlerin türünü belirledikten sonra onları nasıl ele alacağımıza karar verebiliriz: 
1. Eksik Değerleri Silme (Eksiklik Oranı Düşükse) Eğer bir özellikte çok az eksik değer varsa (örn. <%5), sadece o satırları silmek modeli önemli ölçüde etkilemeyebilir.
•	Örnek: df.dropna(inplace=True) 
2. Veri Doldurma (Imputation) 
•	Ortalama/Medyan/Mod ile Doldurma: Eksik değerleri ortalama (sayısal veriler için), medyan (çarpık sayısal veriler için) veya mod (kategorik veriler için) ile değiştirmek. df['Age'].fillna(df['Age'].mean(), inplace=True) # Ortalama Doldurma 
•	İleri veya Geri Doğru Doldurma: Eksik değerleri önceki veya sonraki mevcut değerleri kullanarak doldurma (zaman serisi verileri için iyidir). df.fillna(method='ffill', inplace=True) # İleri doğru doldurma 
•	Tahmine Dayalı Doldurma: Eksik değerleri tahmin etmek için regresyon veya KNN gibi modeller kullanmak.
Python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df[['Age', 'Salary']] = imputer.fit_transform(df[['Age', 'Salary']])
3. Eksiklik Göstergeleri Eklemek (Adding Missing Indicators) Eksik değerleri doldurmak yerine, bir değerin eksik olup olmadığını işaretleyen yeni bir sütun oluşturabiliriz. df['Age_missing'] = df['Age'].isnull().astype(int) 
4. Alan Bilgisi Kullanmak (Using Domain Knowledge) MNAR durumlarında, bazen alan uzmanlığı en iyi yaklaşıma karar vermemize yardımcı olabilir.
•	Örnek: Eksik gelir değerleri yüksek maaşları gösteriyorsa, sadece ortalama ile doldurmak yerine onları "yüksek gelirli" olarak gruplandırabiliriz.
Özet 
Eksik Veri Türü	Örnek	En İyi Ele Alma Yöntemleri
MCAR	Ankette rastgele eksik olan yaş değerleri	Satırları düşürün veya basit doldurma kullanın
MAR	Eksik gelir değerlerinin eğitim düzeyine bağlı olması	Doldurma için regresyon veya ilgili özellikleri kullanın
MNAR	Yüksek kazananların maaşlarını bildirmemesi	Alan bilgisi kullanın veya eksikliği modelleyin
        
Son İpucu: Eksik değerleri nasıl ele alacağınıza karar vermeden önce, eksikliğin ardındaki nedeni anlamak için her zaman verinizi keşfedin. Doğru bir kullanım, makine öğrenmesi modelinizin doğruluğunu önemli ölçüde artırabilir!

Veri Ön İşlemede Aykırı Değerleri İşlemeyi (Outlier Handling) Anlamak (Başlangıç Dostu) 
Aykırı değerler (outliers), verilerinizin geri kalanından önemli ölçüde sapan aşırı uç değerlerdir. Örneğin, bir veri setindeki yaş ortalaması 30 ise, ancak bir kişinin yaşı 150 olarak kaydedilmişse, bu bir aykırı değerdir. Aykırı değerler, özellikle Doğrusal Regresyon (Linear Regression) veya Lojistik Regresyon (Logistic Regression) gibi doğrusal modeller olmak üzere, sonuçları büyük ölçüde çarpıtarak makine öğrenmesi modellerini mahvedebilir.
EDA (Keşifçi Veri Analizi) sırasında aykırı değerleri tespit ettikten sonra (Kutu Grafikleri - Boxplots veya Z-skorları kullanarak), bunları ele almanız gerekir.
Aykırı Değerleri Ele Almak İçin Temel Teknikler 
1. Kırpma (Trimming - Aykırı Değerleri Kaldırma) En basit yaklaşımdır. Kabul edilebilir sınırları hesaplarsınız ve bu sınırların dışına çıkan satırları tamamen düşürürsünüz.
•	Formül (IQR Yöntemi): Alt Sınır (Lower Bound) = Q1 - (1.5 * IQR) | Üst Sınır (Upper Bound) = Q3 + (1.5 * IQR) 
•	Örnek (Python): 
Python
import pandas as pd
import numpy as np

# Q1, Q3 ve IQR'ı hesapla
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Sınırları belirle
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Sadece sınırlar içindeki satırları tut
df_trimmed = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
📌 Ne Zaman Kullanılır: Devasa bir veri setiniz varsa ve aykırı değerler verilerinizin çok küçük bir yüzdesini (örn. %5'ten az) oluşturuyorsa.
2. Kırpma / Winsorizasyon (Capping - Aykırı Değerleri Değiştirme) Aykırı değerleri silmek yerine, onları belirli bir maksimum veya minimum değerle sınırlandırırsınız. Örneğin, 200.000$'ın üzerindeki herhangi bir maaş tam olarak 200.000$ ile değiştirilir.
•	Örnek (Python): 
Python
# Değerleri yukarıda hesaplanan alt ve üst sınırlarda sınırlandır (cap)
df['Salary_Capped'] = np.where(df['Salary'] > upper_bound, upper_bound,
                      np.where(df['Salary'] < lower_bound, lower_bound, df['Salary']))
📌 Ne Zaman Kullanılır: Veri satırlarını kaybetmeyi göze alamadığınız, ancak yine de bu devasa sayıların aşırı etkisini en aza indirmeniz gerektiği durumlarda.
3. Log Dönüşümü (Log Transformation) Büyük sayıları birbirine yaklaştırırken küçük sayıları nispeten aralıklı tutan matematiksel bir numaradır. Sağa çarpık (right-skewed) verileri mükemmel şekilde ele alır.
•	Örnek (Python): 
Python
import numpy as np

# Logaritmik dönüşüm uygula
df['Salary_Log'] = np.log(df['Salary'] + 1) # log(0)'dan kaçınmak için 1 eklendi
📌 Ne Zaman Kullanılır: Verilerinizin "uzun bir kuyruğu" (yüksek derecede çarpık) varsa ve verinin normal, çan eğrisi dağılımına daha yakın görünmesini istiyorsanız.
Aykırı Değerleri Ele Almanın Özeti 
•	Kırpma (Trimming):
o	Artıları: Kötü veriyi tamamen ortadan kaldırır.
o	Eksileri: Küçük veri setleri için kötü olabilecek veri noktalarını kaybedersiniz.
•	Sınırlandırma (Capping/Winsorization):
o	Artıları: Aşırı değerleri etkisiz hale getirirken veri setinin boyutunu olduğu gibi korur.
o	Eksileri: Orijinal veriyi değiştirir ve sınır noktalarında yapay bir "yığılma" yaratır.
•	Log Dönüşümü:
o	Artıları: Yüksek derecede çarpık verileri doğal olarak normalleştirmek için harikadır.
o	Eksileri: Verinin yorumlanmasını zorlaştırır (Örn. Maaş 100.000$ yerine 11.5 olur).
Son İpucu: Tüm modeller aykırı değerlerden korkmaz! Ağaç tabanlı algoritmalar (Karar Ağaçları, Rastgele Ormanlar ve XGBoost gibi) aykırı değerlere karşı inanılmaz derecede dayanıklıdır. Verileriniz kolayca düzeltemeyeceğiniz aşırı değerlerle doluysa, doğrusal bir model yerine ağaç tabanlı bir modele geçmeyi düşünün! 
 
Özellik Mühendisliğinde Veri Kodlamayı (Data Encoding) Anlamak 
Makine öğrenmesi modelleriyle çalışırken, genellikle sayısal değerler yerine etiketlerden veya kategorilerden oluşan verilerle, yani kategorik verilerle karşılaşırız. Ancak, makine öğrenmesi modelleri en iyi sayılarla çalışır, bu nedenle veri kodlama (data encoding) tekniklerini kullanarak kategorik verileri sayısal formata dönüştürmemiz gerekir.
Veri Kodlama Neden Önemlidir? 
•	Makine öğrenmesi modelleri sayısal girdi gerektirir ve kategorik değerler sayılara dönüştürülmelidir.
•	Doğru kodlama, modellerin kategoriler arasındaki ilişkileri anlamasına yardımcı olur.
•	Farklı kategorik veri türleri için farklı kodlama teknikleri uygundur.
Veri Kodlama Türleri Farklı durumlara uygun çok sayıda kodlama tekniği vardır.
1. One-Hot Kodlama (One-Hot Encoding) Her benzersiz kategoriyi ayrı bir ikili (0 veya 1) sütuna dönüştürür. Az sayıda benzersiz değere sahip nominal (sırasız) kategoriler için en iyisidir.
•	Örnek: 
Python
import pandas as pd
df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green']})
df_encoded = pd.get_dummies(df, columns=['Color'])
print(df_encoded)
📌 Ne Zaman Kullanılır: Kategorik değişkenin az sayıda benzersiz değeri varsa.
2. Etiket Kodlama (Label Encoding) Her kategoriye benzersiz bir tam sayı atar. Sıralamanın önemli olduğu ordinal (sıralı) kategoriler (Örn. Düşük, Orta, Yüksek) için uygundur.
•	Örnek: 
Python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Size_Encoded'] = encoder.fit_transform(df['Size'])
📌 Ne Zaman Kullanılır: Kategoriler arasında doğal bir sıra olduğunda (Örn. eğitim seviyeleri, sıralamalar).
3. Sıralı Kodlama (Ordinal Encoding) Etiket Kodlamaya (Label Encoding) benzer, ancak sıralamayı manuel olarak belirlememize olanak tanır.
•	Örnek (Sıralamayı Manuel Atama): 
Python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['High School', "Bachelor's", "Master's"]])
df['Education_Encoded'] = encoder.fit_transform(df[['Education']])
📌 Ne Zaman Kullanılır: Kategoriler anlamlı bir sıralama izlediğinde.
4. Frekans Kodlama (Frequency Encoding) Her kategoriyi veri setindeki frekansı (sayısı) ile değiştirir. Çok sayıda benzersiz kategori olduğunda yardımcı olur.
•	Örnek: 
Python
df['City_Freq'] = df['City'].map(df['City'].value_counts())
📌 Ne Zaman Kullanılır: Kategorinin frekansı önemli bir özellik olabilecekse.
5. Hedef Kodlama (Target / Mean Encoding) Kategorileri hedef değişkenin (target) ortalaması ile değiştirir. Kategorik özelliğin hedef değişkenle güçlü bir korelasyona sahip olduğu durumlar için en iyisidir.
•	Örnek: 
Python
df['Product_Encoded'] = df.groupby('Product')['Sales'].transform('mean')
📌 Ne Zaman Kullanılır: Kategorik özellik hedef değişkeni etkiliyorsa.
6. İkili Kodlama (Binary Encoding) Kategorileri ikili (binary) formata dönüştürür ve bunları ayrı sütunlara kodlar. Yüksek kardinaliteye (çok çeşitli değere) sahip kategorik veriler için One-Hot Kodlamadan daha fazla bellek verimliliği sağlar.
•	Örnek (4 kategori için): 
Python
from category_encoders import BinaryEncoder
encoder = BinaryEncoder(cols=['Category'])
df_encoded = encoder.fit_transform(df)
📌 Ne Zaman Kullanılır: Veri setinde çok sayıda benzersiz kategori (yüksek kardinalite) varsa.
Doğru Kodlama Tekniğini Seçmek 
•	One-Hot Encoding: Az sayıda benzersiz değere sahip nominal kategoriler için. (Artıları: Basit, yaygın | Eksileri: Çok fazla kategori varsa çok fazla sütun yaratır) 
•	Label Encoding: Sıralı kategoriler (Ordinal) için. (Artıları: Verimli, az bellek kullanır | Eksileri: Modelleri, sayıların sıra ifade ettiği konusunda yanıltabilir) 
•	Ordinal Encoding: Sıralaması bilinen kategoriler için. (Artıları: Sıralamaya uyar | Eksileri: Sıralamanın önceden bilinmesini gerektirir) 
•	Frequency Encoding: Çok fazla benzersiz kategorisi olanlar için. (Artıları: Kompakt gösterim | Eksileri: Aşırı öğrenmeye - overfitting'e yol açabilir) 
•	Target Encoding: Hedefle güçlü ilişkisi olan kategoriler için. (Artıları: Hedefle ilişkiyi korur | Eksileri: Veri sızıntısına - data leakage neden olabilir) 
•	Binary Encoding: Yüksek kardinaliteli kategoriler için. (Artıları: Çok kategoride verimlidir | Eksileri: Yorumlanması o kadar kolay değildir) 
Son İpucu: Bir kodlama yöntemi seçmeden önce, verileri analiz edin ve kategorilerin doğasını anlayın. Doğru kodlama tekniği, gereksiz karmaşıklıktan kaçınırken model performansını artırabilir.

Özellik Mühendisliğinde Özellik Ölçeklendirmeyi (Feature Scaling) Anlamak (Başlangıç Dostu)
Makine öğrenmesi modelleri (özellikle KNN gibi uzaklığa dayalı veya Doğrusal Regresyon ve Sinir Ağları gibi gradyan inişine dayalı olanlar) sayıların büyüklüğüne karşı oldukça hassastır. Bir özelliğin değerleri 0 ile 1 arasında değişirken, bir diğerininki 1.000 ile 100.000 arasında değişiyorsa, model yanlışlıkla büyük sayıların daha önemli olduğunu varsayabilir.
Bunu düzeltmek için, temel ilişkilerini değiştirmeden tüm sayısal özellikleri benzer bir ölçeğe getirmek amacıyla Özellik Ölçeklendirme (Feature Scaling) kullanırız.
Özellik Ölçeklendirme Türleri Sayısal verileri ölçeklendirmek için kullanılan üç temel teknik vardır:
1. Standardizasyon (Z-Skoru Ölçeklendirme - Standardization) Veriyi 0 ortalama ve 1 standart sapma etrafında merkezler. Verilerinizin normal dağıldığını (Gauss) varsayar.
•	Formül: z = (x - ortalama) / standart_sapma 
•	Örnek (Python):
Python
from sklearn.preprocessing import StandardScaler
import pandas as pd

df = pd.DataFrame({'Age': [25, 30, 35, 40], 'Salary': [50000, 60000, 80000, 120000]})

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df_scaled)
📌 Ne Zaman Kullanılır: Normal bir dağılım varsayan veya uzaklıklara dayanan algoritmalar (Örn. Lojistik Regresyon, SVM, PCA, KNN) kullanıyorsanız.
2. Normalizasyon (Min-Maks Ölçeklendirme - Normalization) Veriyi, tüm değerler kesinlikle belirli bir aralıkta, genellikle 0 ile 1 arasında olacak şekilde ölçeklendirir.
•	Formül: x_olcekli = (x - x_min) / (x_maks - x_min) 
•	Örnek (Python):
Python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
📌 Ne Zaman Kullanılır: Verileriniz normal bir dağılım izlemiyorsa veya veri dağılımı hakkında varsayımda bulunmayan algoritmalar (Örn. Sinir Ağları, Görüntü İşleme) kullanırken.
3. Güçlü Ölçeklendirme (Robust Scaling) Standardizasyona benzer, ancak ortalama ve standart sapma yerine medyanı (median) ve Çeyrekler Arası Aralığı (IQR) kullanır.
•	Formül: x_robust = (x - medyan) / (Q3 - Q1) 
•	Örnek (Python):
Python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

Ölçeklendirme Türlerinin Özeti
•	Standardizasyon: * Çıktı: Ortalama = 0, Standart Sapma = 1 
o	Artıları: Değişen ölçekleri iyi idare eder, aykırı değerleri korur.
o	Eksileri: Belirli bir aralıkla sınırlı değildir (negatif değerler alabilir).
•	Normalizasyon: * Çıktı: Genellikle kesinlikle [0, 1] arası 
o	Artıları: Tüm değerleri tamamen aynı aralıkta tutar.
o	Eksileri: Aşırı aykırı değerlere karşı son derece hassastır.
•	Güçlü Ölçeklendirme (Robust): * Çıktı: Medyan ve IQR'a dayanır 
o	Artıları: Aykırı değerleri tamamen görmezden gelir.
o	Eksileri: Orijinal dağılım şeklini biraz kaybeder.
Son İpucu:
Veri Sızıntısının Altın Kuralı (Golden Rule of Data Leakage): Eğitim verilerinizde her zaman .fit_transform() kullanın, ancak test/doğrulama verilerinizde SADECE .transform() kullanın. Modeli adil bir şekilde değerlendirmek için ölçekleyici min/maks veya ortalama değerlerini kesinlikle sadece eğitim setinden öğrenmelidir!
 
Makine Öğrenmesinde Özellik Seçimini (Feature Selection) Anlamak (Başlangıç Dostu)
Makine öğrenmesinde daha fazla veri her zaman daha iyi değildir. Çok fazla özelliğe sahip olmak "Boyutluluk Laneti" (Curse of Dimensionality) adı verilen bir soruna yol açabilir. Modelinizi yavaşlatabilir, yorumlanmasını zorlaştırabilir ve aşırı öğrenmeye (overfitting - kalıpları öğrenmek yerine eğitim verilerini ezberlemeye) eğilimli hale getirebilir.
Özellik Seçimi (Feature Selection), işe yaramayan veya gereksiz olanları atarken modeliniz için en alakalı ve önemli özellikleri seçme sürecidir.
Özellik Seçiminin Temel Türleri En iyi özellikleri seçmek için üç temel strateji vardır:
1. Filtre Yöntemleri (Filter Methods) Bu yöntemler, tamamen herhangi bir makine öğrenmesi modelinden bağımsız olarak, her özellik ile hedef değişken arasındaki ilişkiyi değerlendirmek için istatistiksel testler kullanır. Çok hızlıdırlar ve genellikle ilk adımdır.
•	Örnekler: Korelasyon Matrisi (yüksek düzeyde ilişkili özellikleri düşürme), Ki-Kare testi, ANOVA.
•	Örnek (Python - Düşük varyanslı özellikleri düşürme):
Python
from sklearn.feature_selection import VarianceThreshold

# Değerin örneklerin %90'ında veya daha fazlasında aynı olduğu tüm özellikleri kaldırır
selector = VarianceThreshold(threshold=0.1)
df_filtered = selector.fit_transform(df_scaled)
📌 Ne Zaman Kullanılır: Devasa bir veri setiniz varsa ve bir modeli eğitmeden önce bariz "çöp" özellikleri kaldırmak için hızlı, hesaplama açısından ucuz bir yola ihtiyacınız olduğunda.
2. Sarmal Yöntemler (Wrapper Methods) Bu yöntemler, makine öğrenmesi modelini özellikleri farklı kombinasyonlarla gerçekten eğiterek ve performansını kontrol ederek değerlendirir. Son derece doğru ancak hesaplama açısından pahalıdırlar.
•	Örnekler: Özyineli Özellik Eleme (RFE - Recursive Feature Elimination), İleri Seçim (Forward Selection), Geriye Doğru Eleme (Backward Elimination).
•	Örnek (Python - Özyineli Özellik Eleme):
Python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
# En önemli 5 özelliği seç
rfe = RFE(estimator=model, n_features_to_select=5)
df_selected = rfe.fit_transform(X, y)
📌 Ne Zaman Kullanılır: Daha küçük bir veri setiniz ve en iyi seti bulmak amacıyla farklı özellik kombinasyonlarını test etmek için yeterli hesaplama gücünüz olduğunda.
3. Gömülü Yöntemler (Embedded Methods) Bu yöntemler, model eğitimi süreci boyunca özellik seçimini otomatik olarak gerçekleştirir. Algoritmanın kendisi öğrenirken hangi özelliklerin önemli olduğunu çözer.
•	Örnekler: Lasso Regresyonu (L1 Düzenlileştirme - önemsiz özellik ağırlıklarını tam olarak sıfıra zorlar), Ağaç tabanlı Özellik Önemi (Rastgele Orman - Random Forest, XGBoost).
•	Örnek (Python - Ağaç tabanlı Önem):
Python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X, y)

# Her bir özellik için önem puanını görüntüle
importance_scores = model.feature_importances_
print(importance_scores)
📌 Ne Zaman Kullanılır: Rastgele Orman, XGBoost veya Lasso gibi özellik önemini sizin için doğal olarak hesaplayan güçlü algoritmalar kullanırken.
Seçim Yöntemlerinin Özeti
•	Filtre Yöntemleri: * Artıları: Çok hızlıdır, aşırı öğrenmeyi (overfitting) önler, modelden bağımsızdır.
o	Eksileri: Özelliklerin yalnızca diğerleriyle birleştiğinde önemli olduğu ilişkileri kaçırabilir.
•	Sarmal Yöntemler: * Artıları: Özelliklerin olası en iyi kombinasyonunu bulur.
o	Eksileri: Çok yavaştır, küçük veri setlerinde yüksek aşırı öğrenme (overfitting) riski taşır.
•	Gömülü Yöntemler: * Artıları: Doğrudan modele yerleştirilmiştir, hız ve doğruluğu dengeler.
o	Eksileri: Belirli bir algoritmaya bağlıdır (Örn. Rastgele Orman'dan elde edilen özellik önemi, bir SVM modeline mükemmel şekilde uymayabilir).
Son İpucu:
Çöp Giren Çöp Çıkar (Garbage In, Garbage Out)! Modelinizi gürültüden (noise) öğrenmeye zorlamayın. Alakasız özellikleri kaldırmak, model doğruluğunu genellikle modelin karmaşık hiperparametrelerini ayarlamaktan çok daha fazla artırır.
 
Makine Öğrenmesinde Dengesiz Verileri (Imbalanced Data) Anlamak (Başlangıç Dostu)
Birçok gerçek dünya sınıflandırma probleminde, tahmin etmek istediğiniz sınıflar eşit olarak temsil edilmez. Örneğin, dolandırıcılık tespitinde işlemlerin %99'u normal (Çoğunluk Sınıfı - Majority Class) ve sadece %1'i dolandırıcılık (Azınlık Sınıfı - Minority Class) olabilir.
Bir modeli bu veri üzerinde eğitirseniz, her seferinde basitçe "Normal" tahmininde bulunabilir ve %99 doğruluk elde edebilir! Ancak, hiçbir dolandırıcılığı yakalayamadığı için model tamamen işe yaramaz olacaktır. Bu durum Doğruluk Paradoksu (Accuracy Paradox) olarak bilinir.
Bunu düzeltmek için, modelimizi eğitmeden önce veri setini dengeleyecek teknikler kullanırız.
Dengesiz Verileri Ele Almak İçin Temel Teknikler
1. Azaltarak Örnekleme (Undersampling - Çoğunluk Sınıfını Azaltma) Bu teknik, çoğunluk sınıfındaki örnekleri azınlık sınıfının boyutuyla eşleşene kadar rastgele kaldırır.
•	Örnek (Python):
Python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

# X özellikleriniz, y ise hedef değişkeniniz olsun
undersampler = RandomUnderSampler(random_state=42)
X_under, y_under = undersampler.fit_resample(X, y)
📌 Ne Zaman Kullanılır: Devasa bir veri setiniz varsa ve kritik bilgileri kaybetmeden bazı verileri kaybetmeyi göze alabiliyorsanız.
2. Çoğaltarak Örnekleme (Oversampling - Azınlık Sınıfını Çoğaltma) Bu teknik, azınlık sınıfındaki örnekleri çoğunluk sınıfının boyutuyla eşleşene kadar rastgele kopyalar.
•	Örnek (Python):
Python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(random_state=42)
X_over, y_over = oversampler.fit_resample(X, y)
📌 Ne Zaman Kullanılır: Çok küçük bir veri setiniz varsa ve hiçbir veriyi silmeyi göze alamıyorsanız (Undersampling, modeli eğitmek için çok az veri bırakacaktır).
3. SMOTE (Sentetik Azınlık Aşırı Örnekleme Tekniği) SMOTE, azınlık sınıfının tam kopyalarını çıkarmak yerine, mevcut azınlık veri noktalarına benzer yepyeni, sentetik örnekler oluşturmak için bir En Yakın Komşular (Nearest Neighbors) algoritması kullanır.
•	Örnek (Python):
Python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
📌 Ne Zaman Kullanılır: Standart Oversampling'in aşırı öğrenme (overfitting) riskinden kaçınmak istediğinizde. SMOTE, dengesiz veriler için genellikle en popüler ve etkili yöntemdir.
Dengeleme Yöntemlerinin Özeti
•	Undersampling: * Artıları: Veri seti küçüleceği için eğitimi çok daha hızlı hale getirir.
o	Eksileri: Çoğunluk sınıfındaki potansiyel olarak faydalı verileri ve kalıpları çöpe atmış olursunuz.
•	Oversampling: * Artıları: Hiçbir bilgiyi kaybetmezsiniz.
o	Eksileri: Tamamen aynı veri noktalarını kopyalamak, modelin bunları ezberlemesine (Aşırı Öğrenme) neden olabilir.
•	SMOTE: * Artıları: Modelin daha önce görülmemiş verilerde daha iyi genelleme yapmasına yardımcı olan çeşitli, yeni veri noktaları oluşturur.
o	Eksileri: Hesaplanması daha yavaştır; azınlık noktaları çoğunluk noktalarıyla örtüşüyorsa bazen gürültülü veriler oluşturabilir.
Son İpucu:
Veri setinizin tamamını asla bölmeden önce dengelemeyin ! Her zaman ÖNCE verilerinizi Eğitim (Training) ve Test setlerine ayırın. Ardından, SMOTE veya Oversampling'i SADECE Eğitim Setine uygulayın. Bölmeden önce uygularsanız, sentetik veriler test setinize sızacak ve model değerlendirmeniz tamamen sahte olacaktır! Ayrıca, dengesiz veri setlerini değerlendirmek için asla "Doğruluk (Accuracy)" kullanmayın — her zaman Kesinlik (Precision), Duyarlılık (Recall) ve F1-Skoru (F1-Score) kullanın.
 
Veri Bölme (Data Splitting) ve Çapraz Doğrulamayı (Cross-Validation) Anlamak (Başlangıç Dostu)
Bir öğrenciyi sınava hazırlarken test sorularını önceden gösterirseniz %100 alır, ancak aslında hiçbir şey öğrenmemiştir; sadece cevapları ezberlemiştir.
Aynı şey makine öğrenmesi için de geçerlidir. Bir modeli tüm veri setiniz üzerinde eğitir ve ardından tam olarak aynı veri seti üzerinde değerlendirirseniz, kusursuz görünecektir (Aşırı Öğrenme - Overfitting), ancak gerçek dünyada sefil bir şekilde başarısız olacaktır. Bunu önlemek için, eğitimden önce verilerimizi farklı setlere bölmeliyiz.
Modelleri Değerlendirmek İçin Temel Teknikler
1. Eğitim / Test Bölünmesi (Train / Test Split) Veri setinizi rastgele iki parçaya bölersiniz: modelin öğrenmesi için büyük bir parça (Eğitim - Train) ve daha sonra üzerinde test etmek için daha küçük, gizli bir parça (Test). Yaygın bir oran %80 Eğitim ve %20 Test şeklindedir.
•	Örnek (Python):
Python
from sklearn.model_selection import train_test_split

# Veriyi böl: %80 eğitim için, %20 test için
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
📌 Ne Zaman Kullanılır: Yeterince büyük bir veri setiniz varsa ve modelinizi değerlendirmek için hızlı, standart bir yola ihtiyacınız olduğunda.
2. Eğitim / Doğrulama / Test Bölünmesi (Train / Validation / Test Split) Modelinizin ayarlarını (Hiperparametre Ayarı - Hyperparameter Tuning) değiştiriyorsanız, üçüncü bir sete ihtiyacınız vardır.
•	Eğitim Seti (Train Set): Modele öğretmek için.
•	Doğrulama Seti (Validation Set): Farklı ayarları test etmek ve en iyi modeli seçmek için.
•	Test Seti (Test Set): Nihai performansı kontrol etmek için tamamen dokunulmamış nihai veri. 📌 Ne Zaman Kullanılır: Karmaşık modelleri (Derin Öğrenme veya XGBoost gibi) optimize ederken ve yanlışlıkla test setinize aşırı öğrenmekten kaçınmak istediğinizde.
3. K-Katlamalı Çapraz Doğrulama (K-Fold Cross-Validation) Eğitim/test bölünmesi rastgele bir kesime dayanır. Peki ya tüm zor örnekler rastgele test setinize düşerse? K-Katlamalı Çapraz Doğrulama, verileri "K" eşit parçaya (Örn. 5 parça) bölerek bunu düzeltir. Modeli, her seferinde farklı bir parçayı test seti, geri kalanını ise eğitim seti olarak kullanarak 5 kez eğitir. Son olarak, puanların ortalamasını alır.
•	Örnek (Python):
Python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
# 5 Katlamalı Çapraz Doğrulama Gerçekleştir
scores = cross_val_score(model, X, y, cv=5)

print("Average Accuracy:", scores.mean())
📌 Ne Zaman Kullanılır: Veri setiniz küçükse. Çapraz doğrulama, her bir veri noktasının test setinde yer alma şansı elde etmesini sağlayarak size çok daha güvenilir bir doğruluk puanı verir.
Veri Bölme Yöntemlerinin Özeti
•	Train/Test Split: * Artıları: Basit, hızlı, standart.
o	Eksileri: Rastgele bölünmeye büyük ölçüde bağlıdır; küçükse veri setinin tamamını iyi temsil etmeyebilir.
•	Train/Val/Test Split: * Artıları: Hiperparametre ayarı için mükemmeldir.
o	Eksileri: Gerçek eğitim için mevcut veri miktarını azaltır.
•	K-Fold Cross-Validation: * Artıları: En sağlam ve güvenilir değerlendirme metriğidir, veri kullanımını en üst düzeye çıkarır.
o	Eksileri: Hesaplama açısından pahalıdır (K=5 ise, 5 farklı model eğitiyorsunuz demektir).
Son İpucu:
Ön İşlemenin Nihai Kuralı: Herhangi bir Ölçeklendirme (Scaling), Eksik Değer Doldurma (ortalama kullanmak gibi) veya SMOTE uygulamadan ÖNCE verilerinizi Eğitim ve Test setlerine ayırmak ZORUNDASINIZ. Bu adımları bölmeden önce tüm veri setine uygularsanız, test setinden gelen bilgiler eğitim setine "sızar" (Veri Sızıntısı - Data Leakage) ve modelinizin bütünlüğünü mahveder!
